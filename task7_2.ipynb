```python
listed_items = input()
```

```python
def check_input(slot):  #  проверяет элемент и возвращает True, если строка
    return type(slot) == str
```

```python
def get_substring(search_substr): #  возвращает самую длинную подстроку без повторений в переданной строке
    result, temporary = '', ''

    for element in search_substr:
        if element in temporary:
            temporary = temporary[temporary.index(element) + 1:]
        temp += element

        if len(temporary) > len(result):
            result = temporary

    return result
```

```python
def main(listed_items):  #  принимает список, находит самую длинную строку, передает на метод get_substring, печатает на экран индекс самой длинной строки и подстроку
    index, longest = -1, ""

    for i in range(len(listed_items)):
        if check_input(listed_items[i]) and len(listed_items[i]) > len(longest):
            longest = listed_items[i]
            index = i

    if index == -1:
        print(-1, "")
        return

    print(index, get_substring(longest))
```

```python
def check_input(slot):
    return type(slot) == str
```

```python
def get_substring(s):
    result = ""
    current = ""

    for char in s:
        if char in current:
            current = current[current.index(char) + 1:]
        current += char

        if len(current) > len(result):
            result = current

    return result
```

```python
def main(arr):
    max_index = -1
    max_string = ""

    for i, item in enumerate(arr):
        if check_input(item):
            if max_index == -1 or len(item) > len(max_string):
                max_string = item
                max_index = i

    if max_index == -1:
        print(-1, "")
        return

    print(f"{max_index}, '{get_substring(max_string)}'")
```
